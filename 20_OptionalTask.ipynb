```python
df['brand'] = df['brand'].fillna('Unknown')
df['type'] = df['type'].fillna('Unknown')
df['available'] = df['available'].fillna(df['available'].median())
df['availableText'] = df['availableText'].fillna('Unknown')
df['sold'] = df['sold'].fillna(df['sold'].median())
df['lastUpdated'] = df['lastUpdated'].fillna('Unknown')
```

```python
df.describe()
```

```python
input_X = df.drop('price', axis=1)
output_y = df['price']
input_X.shape,output_y.shape
```

```python
encoded_input_X = pd.get_dummies(input_X)
encoded_input_X
```

```python
encoded_input_X = encoded_input_X.astype('int')
encoded_input_X
```

```python
numeric_col = ['available','sold']
scaler = StandardScaler()
encoded_input_X[numeric_col] = scaler.fit_transform(encoded_input_X[numeric_col])
encoded_input_X
```

```python
X_train,X_test,y_train,y_test = train_test_split(encoded_input_X,output_y,test_size=0.20,random_state=42)
X_train
```

```python
X_test
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train,y_train)
```

```python
model_lr.predict(X_test)
```

```python
y_test
```

```python
r2_score(y_test,model_lr.predict(X_test))
```
