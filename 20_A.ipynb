```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
ds = pd.read_csv("ford_car_dataset.csv")
```

```python
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds.describe()
```

```python
ds.info()
```

```python
ds.duplicated().sum()
```

```python
ds.drop_duplicates(inplace=True)
```

```python
ds.duplicated().sum()
```

```python
X = ds.drop("price", axis=1)
Y = ds["price"]
X.shape
```

```python
Y.shape
```

```python
X = pd.get_dummies(X, dtype=int)

print(X.head())
```

```python
scaler = StandardScaler()

num_cols = ["year", "mileage", "tax", "mpg", "engineSize"]

X[num_cols] = scaler.fit_transform(X[num_cols])

print(X.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=0.33,random_state=42)
X_train
X_test
```

```python
y_train
y_test
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

print("Intercept :", lr.intercept_)
print("Coefficients :")
print(lr.coef_)
```

```python
y_pred = lr.predict(X_test)

print("Predicted Values")
print(y_pred[:10])

print("\nActual Values")
print(y_test.iloc[:10])
```

```python
score = r2_score(y_test, y_pred)

score
```

The R² score tells us how well the model predicts car prices. A value closer to 1 means better performance. If the score is high, the model is able to explain most of the variation in the dataset.

```python
import joblib

joblib.dump(lr, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "columns.pkl")
```
